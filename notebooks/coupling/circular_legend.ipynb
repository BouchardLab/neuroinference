```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Semicircle

```python
xval = np.arange(0, 1.0 * np.pi, 0.01)
yval = np.ones_like(xval)

colormap = plt.get_cmap('twilight')
norm = matplotlib.colors.Normalize(0.0, np.pi)

ax = plt.subplot(1, 1, 1, polar=True)
ax.scatter(xval, yval,
           c=xval, s=50,
           marker='o',
           cmap=colormap,
           norm=norm,
           linewidths=0)
ax.set_xticks([])
ax.set_yticks([])

for spine in ax.spines:
    ax.spines[spine].set_visible(False)

plt.savefig('semicircle.pdf', bbox_inches='tight', transparent=True)
```

## Full Circle

```python
xval = np.arange(0, 2.0 * np.pi, 0.01)
yval = np.ones_like(xval)

colormap = plt.get_cmap('twilight')
norm = matplotlib.colors.Normalize(0.0, 2 * np.pi)

ax = plt.subplot(1, 1, 1, polar=True)
ax.scatter(xval, yval,
           c=xval, s=50,
           marker='o',
           cmap=colormap,
           norm=norm,
           linewidths=0)
ax.set_xticks([])
ax.set_yticks([])

for spine in ax.spines:
    ax.spines[spine].set_visible(False)
    
plt.savefig('fullcircle.pdf', bbox_inches='tight', transparent=True)
```
